```python
# import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')
```

Load dataset

```python
pd.set_option("display.max_columns", None)
df = pd.read_csv("Superstore.csv", encoding='latin1')
df.head(10)
```

Data Overview

```python
df.info()
```

```python

df.describe()
```

```python

df.dtypes
```

Data Type Conversion

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%m/%d/%Y')
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')
```

Chech Missing Values

```python
df.isna().sum()
```

Introduce Missing Values

```python
df2 = df.copy()
df2.loc[0:5, 'Sales'] = np.nan
```

Remove Records

```python
df_remove = df2.dropna()

print("df with NaN:", df2.shape)
print("df without NaN:", df_remove.shape)

df_remove.isna().sum()
```

Mean Imputation

```python
df_mean = df2.copy()
df_mean['Sales'] = df_mean['Sales'].fillna(df_mean['Sales'].mean())

df_mean.head()
```

Median Imputation

```python
df_median = df2.copy()
df_median['Sales'] = df_median['Sales'].fillna(df_median['Sales'].median())

df_median.head()
```

Boxplot Visulization

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Sales'])
plt.title("Boxplot of Sales")
plt.show()
```

```python

plt.figure(figsize=(10,4))
sns.boxplot(x=np.log(df2['Sales']))
plt.show()
```

Detect Outliers

```python
Q1 = df['Sales'].quantile(0.25)
Q3 = df['Sales'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['Sales'] < lower) | (df['Sales'] > upper)]
outliers.head()
```

Remove Outliers

```python
df_no_outliers = df[(df['Sales'] >= lower) & (df['Sales'] <= upper)]

print("Before:", df.shape)
print("After:", df_no_outliers.shape)
```

Capping Outliers

```python
lower_cap = df['Sales'].quantile(0.05)
upper_cap = df['Sales'].quantile(0.95)

df_capped = df.copy()
df_capped['Sales'] = df_capped['Sales'].clip(lower_cap, upper_cap)
```

Boxplot After Capping

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df_capped['Sales'])
plt.title("After Capping")
plt.show()
```

Min-Max Normalization

```python
df_minmax = df[['Sales', 'Quantity']].copy()

df_minmax = (df_minmax - df_minmax.min()) / (df_minmax.max() - df_minmax.min())
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df_scaled = df[['Sales', 'Quantity']].copy()

df_scaled[['Sales', 'Quantity']] = scaler.fit_transform(df_scaled)

df_scaled.head()
```

Z-Score Normalization

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df_standardized = df[['Sales', 'Quantity']].copy()
df_standardized[['Sales', 'Quantity']] = scaler.fit_transform(df_standardized)
```

Heatmap

```python
plt.figure(figsize=(6,4))
sns.heatmap(df_standardized.corr(), annot=True, cmap="coolwarm")
plt.show()
```

Apply PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

df_pca = pca.fit_transform(df_standardized)

df_pca = pd.DataFrame(df_pca, columns=['PC1', 'PC2'])

df_pca.head()
```

PCA Visualization

```python
plt.figure(figsize=(6,4))
plt.scatter(df_pca.iloc[:,0], df_pca.iloc[:,1])
plt.title("PCA Projection")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```
